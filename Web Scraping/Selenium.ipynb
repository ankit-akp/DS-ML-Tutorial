## Find all jobs

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
driver=webdriver.Chrome(executable_path='/usr/bin/chromedriver')

driver.get('http://kanview.ks.gov/PayRates/PayRates_Agency.aspx')
```

```python
a = driver.find_element_by_link_text('Adjutant General')
a.click()

from bs4 import BeautifulSoup
import pandas as pd

dataFrameList = []

for i in range(2) : 
    link = driver.find_element_by_id('MainContent_uxLevel2_JobTitles_uxJobTitleBtn_' + str(i))
    link.click()
    
    data = BeautifulSoup(driver.page_source, 'lxml')
    emp_table = data.table
    print(type(emp_table))
    df = pd.read_html(str(emp_table), header = 0)
    dataFrameList.append(df[0])
    driver.back()
```

```python
a = driver.find_element_by_link_text('Board of Accountancy')
a.click()
```

```python
from selenium.common.exceptions import NoSuchElementException

dataFrameList = []

i = 0
while True : 
    try : 
        link = driver.find_element_by_id('MainContent_uxLevel2_JobTitles_uxJobTitleBtn_' + str(i))
        link.click()

        data = BeautifulSoup(driver.page_source, 'lxml')
        emp_table = data.table
        print(type(emp_table))
        df = pd.read_html(str(emp_table), header = 0)
        dataFrameList.append(df[0])
        driver.back()
        i += 1
    except NoSuchElementException :
        break
```

```python
df = pd.concat(dataFrameList, ignore_index = True)
```

```python
# Implicit Wait

from selenium import webdriver
from selenium.webdriver.common.by import By
driver=webdriver.Chrome(executable_path='/usr/bin/chromedriver')

driver.get('https://www.dell.com/en-in')
# driver.implicitly_wait(10)
text_box=driver.find_element(By.ID,'mh-search-input')
text_box.send_keys('Gaming Laptop')
submit_btn=driver.find_element(By.CLASS_NAME,'mh-search-submit')
submit_btn.click()
driver.implicitly_wait(10)
lap1=driver.find_element(By.XPATH,'//section[contains(@class,"ps-top")]/h3/a')
lap1.click()
# If we provide wait time as 20 secs, and our element is being loaded at 5 sec,
# but still it will wait for 20 sec

# Implicit wait is applied globally mean once I called driver object
# with implicitly_wait() then after that whenever I use driver object
# it will wait for specified amount of secs.
```

# Explicit Wait Conditions




```python
# Explicit waits
# Waits till given condition does not met

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver=webdriver.Chrome(executable_path='/usr/bin/chromedriver')

driver.get('https://www.dell.com/en-in')

text_box=driver.find_element(By.ID,'mh-search-input')
text_box.send_keys('Gaming Laptop')
submit_btn=driver.find_element(By.CLASS_NAME,'mh-search-submit')
submit_btn.click()

# Explicit wait
wait=WebDriverWait(driver,10) # Arguments are driver and upper limit
# Here upper limit is 10 sec
lap1=wait.until(EC.presence_of_element_located((By.XPATH,'//section[contains(@class,"ps-top")]/h3/a')))

lap1.click()
```

```python
# Infinite Scrolling: Scrolling untill data stop coming
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.select import Select
import time
from selenium.webdriver.common.keys import Keys
import selenium.common.exceptions
driver=webdriver.Chrome(executable_path='/usr/bin/chromedriver')

driver.get('https://twitter.com/StackOverheards')
```

```python
currentHeight=driver.execute_script('return document.body.scrollHeight;')
while True:
    driver.execute_script('window.scrollTo(0,arguments[0])',currentHeight)
    time.sleep(3)
    newHeight=driver.execute_script('return document.body.scrollHeight;')
    
    print(currentHeight,newHeight)
    if newHeight==currentHeight:
        break
    currentHeight=newHeight
```
