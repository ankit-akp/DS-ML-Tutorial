```python
from selenium import webdriver
```

```python
driver=webdriver.Chrome(executable_path='/usr/bin/chromedriver')
print(driver)
```

```python
# Visit to given url
# driver.get(url)

driver.get('https://www.codingninjas.com/')
```

```python
# Get title
print(driver.title)
```

```python
# Click back button
driver.back()
```

```python
# Click forward button
driver.forward()
```

```python
# Take screenshot of current page
# driver.save_screenshot('nameofss.extenstion')

driver.save_screenshot('test.png')
```

```python
# How to get html of current page
driver.page_source
```

```python
# How to get link element by knowing the text
from selenium.webdriver.common.by import By
blog_btn=driver.find_element(By.LINK_TEXT,'Blog')
print(blog_btn)
```

```python
# Click on btn
# element.click()
blog_btn.click()
```

```python
# Partial Text
driver.get('https://www.google.com/')
btn=driver.find_element(By.PARTIAL_LINK_TEXT,'ges')
print(btn)
```

```python
btn.click()
```

```python
driver.back()
```

```python
driver.close()
```
