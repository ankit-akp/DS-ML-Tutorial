```python
# More customization
import numpy as np
x=np.arange(50)
y=x**3
plt.plot(x,y,color="black",marker='o') # Here we get dot with lines
plt.show()
```

```python
# Linewidth
plt.plot(x,y,color="black",linewidth=5)
plt.show()
```

```python
# Labels,legend and title
x=np.arange(0,5,0.1)
y1=x**3
y2=x**2
plt.plot(x,y1,color="black",linewidth=5,label='x^3') # label parameter add legend for this graph
plt.plot(x,y2,color="red",linewidth=5,label='x^2') 
plt.xlabel('X')
plt.ylabel('Y')
plt.legend() # This will show legend
plt.title('Graph of X^3')
plt.show()
```

```python
# How to change axis values
import numpy as np
x=np.arange(0,5,0.1)
y=x**3
plt.plot(x,y,color="black",marker='o')
plt.axis([0,10,0,10]) # This list contains for values(xStarting,xEnding,yStarting,yEnding)
plt.show()
```

```python
# How to add grids

import numpy as np
x=np.arange(0,5,0.1)
y=x**3
plt.plot(x,y,color="black",marker='o')
plt.grid()
plt.show()
```

```python
# How to add random text at some place

import numpy as np
x=np.arange(0,5,0.1)
y=x**3
plt.plot(x,y,color="black",marker='o')
plt.grid()
plt.text(2,80,"Hello",fontsize=15) # plt.text(xCoordinate,yCoordinate,text)
plt.show()
```

```python
# Bubble graph
year=[2012,2013,2014,2015,2016,2017]
salary=[12,13,14,17,19,20]
population=[100,120,180,250,300,370]
colors = np.random.rand(6)
plt.scatter(year,salary,s=population,c=colors,alpha=0.5) # Alpha define opacity. Default value is 1
plt.show()
```

```python
# Add edge color in bubble graph
year=[2012,2013,2014,2015,2016,2017]
salary=[12,13,14,17,19,20]
population=[100,120,180,250,300,370]
colors = np.random.rand(6)
plt.scatter(year,salary,s=population,c=colors,edgecolor='black') # edgecolor -> Color of border of bubble
plt.show()
```

```python
# Pie charts

labels=['A','B','C','D']
sizes=[3,4,6,2]
explode=[0.1,0,0,0]
plt.pie(sizes, explode=explode,labels=labels,autopct='%.1f%%',counterclock=False,startangle=100)
plt.axis('equal') # This make pie chart circular
plt.show()
print()
```

```python
# Histograms

a=[1,2,3,4,1,3,4,2,1,1]
plt.hist(a)
plt.show()
```

```python
a = [1, 2, 1, 3, 5, 4, 5, 5, 4, 3, 3, 10, 7, 8, 8, 6, 7, 20, 22, 11, 14, 15]
plt.xticks(np.arange(23))
plt.hist(a, bins = 21, edgecolor = 'black')
plt.show()
```

If bins is an integer, it defines the number of equal-width bins in the range.

If bins is a sequence, it defines the bin edges, including the left edge of the first bin and the right edge of the last bin; in this case, bins may be unequally spaced. All but the last (righthand-most) bin is half-open. In other words, if bins is:

[1, 2, 3, 4]

then the first bin is [1, 2) (including 1, but excluding 2) and the second [2, 3). The last bin, however, is [3, 4], which includes 4.

If bins is a string, it is one of the binning strategies supported by numpy.histogram_bin_edges: 'auto', 'fd', 'doane', 'scott', 'stone', 'rice', 'sturges', or 'sqrt'.
