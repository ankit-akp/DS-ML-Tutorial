```python
from sklearn import datasets
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
import numpy as np
import pandas as pd
```

```python
# Loading data
df=pd.read_csv('training_titanic_x_y_train.csv')

print(df.shape)
df.head(20)
```

```python
# print(df.head(10))
# print("after")
# print(df['Age'].mean())

# print(df.head(10))
df.describe()
print(pd.unique(df['Embarked']))
```

```python
def cleaningData(df):
    df.drop('Name',axis=1,inplace=True)
    df.drop('Ticket',axis=1,inplace=True)
    df.drop('Cabin',axis=1,inplace=True)
#     df.drop('Embarked',axis=1,inplace=True)
    # handling NAN in ages
    df['Age'].fillna(df['Age'].mean(), inplace = True)
    
    def adjustEmbarked(s):
        if s=='S':
            return 0
        elif s=='C':
            return 1
        else:
            return 2

    def func(s):
        if s == 'male':
            return 0
        else:
            return 1

    df['Sex'] = df.Sex.apply(func)
    df['Embarked']=df.Embarked.apply(adjustEmbarked)
cleaningData(df)
df.head(10)
```

```python
df.shape
```

```python
x_train=df.iloc[:,:7]
y_train=df.iloc[:,7]
```

```python
# df.drop('Embarked',axis=1,inplace=True)
df.head(10)
```

```python
print(x_train.shape,y_train.shape)
```

```python
clf = LogisticRegression(C=10, solver = 'liblinear',max_iter=10000)
# clf.fit(x_train, y_train)
```

```python
clf.fit(x_train,y_train)
```

```python
print("Training Score", clf.score(x_train, y_train))
```
