```python
import numpy as np
import pandas as pd
data = np.loadtxt("training_boston_x_y_train.csv", delimiter=",")
x_train=data[:,:13]
y_train=data[:,13]
y_train.shape
```

```python
li = ['x1','x2','x3','x4','x5','x6','x7','x8','x9','x10','x11','x12','x13']
df = pd.DataFrame(x_train)
df.columns = li
df
```

```python
for i in range(13):
    for j in range(i, 13):
        ele = li[i] + "_" + li[j]
        df[ele] = df[li[i]]*df[li[j]]
df.shape
```

```python
x=df.iloc[:,:]
x
```

```python
from sklearn.ensemble import GradientBoostingRegressor
model = GradientBoostingRegressor()

model.fit(x,y_train)
```

```python
## test_boston_x_test.csv
data = np.loadtxt("test_boston_x_test.csv", delimiter=",")
x_test=data[:,:]

x_test.shape
```

```python
li = ['x1','x2','x3','x4','x5','x6','x7','x8','x9','x10','x11','x12','x13']
df = pd.DataFrame(x_test)
df.columns = li
df
```

```python
for i in range(13):
    for j in range(i, 13):
        ele = li[i] + "_" + li[j]
        df[ele] = df[li[i]]*df[li[j]]
df.shape
```

```python
y_test_pred = model.predict(df.iloc[:,:])
y_test_pred
```

```python
score_test=model.score(x,y_train)
score_test
```

```python
np.savetxt('boston_gradient_results.csv', y_test_pred.reshape(-1,1),fmt='%.5f', delimiter =', ')
```
